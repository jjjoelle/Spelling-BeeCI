```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
data_dir = 'DataCollection/outputs/'
ssvep_dir = data_dir + 'SSVEP/'
mi_dir = data_dir + 'MI/'
```

```python
ssvep_files = []
for file in sorted(os.listdir(ssvep_dir)):
    ssvep_files.append(ssvep_dir + file)
mi_files = []
for file in sorted(os.listdir(mi_dir)):
    ssvep_files.append(mi_dir + file)
```

```python
df = pd.read_csv(ssvep_files[0],header=None)
col_names = ['Fp1', 'O1', 'O2','C3','C4','P3','P4','Pz']
df.columns = col_names
```

```python
df
```

```python
def getFileData(fileName):
    df = pd.read_csv(fileName,header=None)
    df.columns = col_names
    return df
def plotChannel(df, channel):
    chan_data = list(df[channel])
    times = np.arange(len(chan_data))*dt
    plt.plot(times, chan_data)
```

```python
Fs = 250
dt = 1/Fs
times = np.arange(1250)*dt
chan_data = list(df['O1'])
plt.plot(times, chan_data);
plt.xlim(2,2.25)
```

```python
df10 = getFileData(ssvep_files[0])
plotChannel(df10, 'Pz')
```

```python
chan_data = list(df['O1'])
ps = np.fft.fft(chan_data)
freqs = np.fft.fftfreq(n=len(chan_data), d=dt)
plt.plot(freqs,ps)
#plt.xlim(-5,5)
```
