```python
import numpy as np
from scipy.signal import lfilter, periodogram

# Generate sample EEG data (10 Hz signal with white noise)
fs = 250 # Sampling rate
t = np.arange(0, 5, 1/fs) # Time vector
eeg_signal = np.sin(2*np.pi*10*t) + 0.1*np.random.randn(len(t))

# AR spectral estimation
order = 30 # AR model order
ar_coeffs = lfilter([1], np.hstack((1, -np.random.rand(order-1))), eeg_signal) # AR coefficients
freqs, power_spectrum = periodogram(eeg_signal, fs=fs, window='hamming', nfft=2**14, detrend=False, return_onesided=True, scaling='density')

# Plot results
import matplotlib.pyplot as plt
plt.semilogy(freqs, power_spectrum, label='Welch periodogram')
plt.semilogy(freqs, np.abs(1/np.polyval(np.flipud(ar_coeffs), np.exp(2j*np.pi*freqs/fs)))**2, label='AR periodogram')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (V^2/Hz)')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.signal import periodogram, arburg

# Generate sample EEG data (10 Hz signal with white noise)
fs = 250 # Sampling rate
t = np.arange(0, 5, 1/fs) # Time vector
eeg_signal = np.sin(2*np.pi*10*t) + 0.1*np.random.randn(len(t))

# AR spectral estimation
order = 30 # AR model order
ar_coeffs, noise_var = arburg(eeg_signal, order=order) # AR coefficients and noise variance
freqs, power_spectrum_ar = periodogram(eeg_signal, fs=fs, window='hamming', nfft=2**14, detrend=False, return_onesided=True, scaling='density')
ar_periodogram = noise_var / np.abs(np.polyval(np.flipud(ar_coeffs), np.exp(-2j*np.pi*freqs/fs)))**2

# Plot results
import matplotlib.pyplot as plt
plt.semilogy(freqs, power_spectrum_ar, label='Welch periodogram')
plt.semilogy(freqs, ar_periodogram, label='AR periodogram')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (V^2/Hz)')
plt.legend()
plt.show()
```
